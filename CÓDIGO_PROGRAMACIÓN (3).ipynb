```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Datos del puente
nombre_puente = "Puente Arroyo Seco"
longitud_puente = 40  # en metros
ancho_puente = 10  # en metros
altura_puente = 4  # en metros
capacidad_agua = longitud_puente * ancho_puente * altura_puente  # m³

# Información básica
print(f"Nombre del Puente: {nombre_puente}")
print(f"Dimensiones del Puente: Longitud {longitud_puente} m, Ancho {ancho_puente} m, Altura {altura_puente} m")
print(f"Capacidad de Almacenamiento de Agua: {capacidad_agua} m³")

# Configuración del gráfico
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Coordenadas del puente (base superior)
x_base = [0, longitud_puente, longitud_puente, 0]
y_base = [0, 0, ancho_puente, ancho_puente]
z_base = [altura_puente] * 4

# Coordenadas del agua (cubo debajo del puente)
x_agua = [0, longitud_puente, longitud_puente, 0, 0]
y_agua = [0, 0, ancho_puente, ancho_puente, 0]
z_agua = [0, 0, 0, 0, altura_puente]

# Dibujar la base del puente
ax.add_collection3d(Poly3DCollection([list(zip(x_base, y_base, z_base))], color='brown', alpha=0.8, label='Base del Puente'))

# Dibujar el agua
verts = [
    [(0, 0, 0), (longitud_puente, 0, 0), (longitud_puente, ancho_puente, 0), (0, ancho_puente, 0)],  # base
    [(0, 0, altura_puente), (longitud_puente, 0, altura_puente), (longitud_puente, ancho_puente, altura_puente), (0, ancho_puente, altura_puente)],  # top
    [(0, 0, 0), (0, ancho_puente, 0), (0, ancho_puente, altura_puente), (0, 0, altura_puente)],  # side 1
    [(longitud_puente, 0, 0), (longitud_puente, ancho_puente, 0), (longitud_puente, ancho_puente, altura_puente), (longitud_puente, 0, altura_puente)],  # side 2
    [(0, 0, 0), (longitud_puente, 0, 0), (longitud_puente, 0, altura_puente), (0, 0, altura_puente)],  # side 3
    [(0, ancho_puente, 0), (longitud_puente, ancho_puente, 0), (longitud_puente, ancho_puente, altura_puente), (0, ancho_puente, altura_puente)],  # side 4
]
ax.add_collection3d(Poly3DCollection(verts, color='cyan', alpha=0.5, label='Agua'))

# Configurar ejes
ax.set_title("Puente Arroyo Seco con Agua Debajo", fontsize=16)
ax.set_xlabel("Longitud (m)")
ax.set_ylabel("Ancho (m)")
ax.set_zlabel("Altura (m)")
ax.set_xlim(0, longitud_puente + 10)
ax.set_ylim(0, ancho_puente + 10)
ax.set_zlim(0, altura_puente + 5)
ax.legend()

plt.show()
```

# **CÁLCULO DEL VOLUMEN DE AGUA BAJO EL PUENTE**

```python
import numpy as np
import matplotlib.pyplot as plt

# Dimensiones del puente
longitud_puente = 40  # en metros
ancho_puente = 10  # en metros
altura_maxima = 4  # en metros

# Función para calcular el volumen de agua
def calcular_volumen(altura):
    if altura > altura_maxima:
        return longitud_puente * ancho_puente * altura_maxima
    return longitud_puente * ancho_puente * altura

# Calcular el volumen de agua con la altura máxima del puente
volumen_maximo = calcular_volumen(altura_maxima)

# Mostrar el volumen máximo de agua que se puede almacenar
print(f"El volumen máximo de agua que se puede almacenar debajo del puente es: {volumen_maximo:.2f} m³")

# Graficar la representación del volumen de agua
alturas = np.linspace(0, altura_maxima, 100)
volumenes = [calcular_volumen(h) for h in alturas]

# Graficar el volumen de agua
plt.figure(figsize=(10, 6))
plt.plot(alturas, volumenes, label="Volumen de Agua", color="blue")
plt.axvline(x=altura_maxima, color="red", linestyle="--", label=f"Altura Máxima ({altura_maxima} m)")
plt.fill_between(alturas, volumenes, color='blue', alpha=0.2)  # Rellenar debajo de la curva
plt.title("Capacidad de Almacenamiento de Agua Bajo el Puente")
plt.xlabel("Altura del Agua (m)")
plt.ylabel("Volumen de Agua (m³)")
plt.legend()
plt.grid()
plt.show()
```

# **ELEVACIÓN DEL TERRENO**

```python
import requests

# Coordenadas del puente
latitud = 19.29137
longitud = -103.79759

# Usar una API para obtener la elevación (requiere clave de API)
def obtener_elevacion(lat, lon):
    try:
        url = f"https://api.opentopodata.org/v1/test-dataset?locations={lat},{lon}"
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            return data['results'][0]['elevation']
        else:
            return "Error al obtener elevación"
    except Exception as e:
        return str(e)

# Obtener elevación
elevacion = obtener_elevacion(latitud, longitud)
print(f"La elevación en las coordenadas ({latitud}, {longitud}) es de {elevacion} metros.")
```

# **CÓDIGO PARA SIMULAR LA RESISTENCIA DEL PUENTE A CRECIDAS Y TERREMOTOS**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del puente
longitud_puente = 40  # en metros
ancho_puente = 10  # en metros
altura_puente = 4  # en metros
peso_material = 2500  # kg/m^3 (peso promedio del concreto)
fuerza_maxima_resistencia = 200000  # N (resistencia máxima del puente a carga externa)

# Parámetros de crecida del río (simulación)
caudal_maximo = 1000  # m³/s (caudal máximo esperado)
densidad_agua = 1000  # kg/m³ (densidad del agua)
presion_agua = caudal_maximo * densidad_agua * 9.81  # Fuerza de presión que el agua ejerce sobre el puente

# Parámetros de terremoto (simulación simple)
intensidad_terremoto = 0.2  # Intensidad de las fuerzas sísmicas (escala de 0 a 1)
longitud_onda_terremoto = 30  # en metros (longitud de onda de las vibraciones)

# Función para calcular el impacto de la crecida del río
def calcular_impacto_crecida(caudal_maximo, densidad_agua):
    return caudal_maximo * densidad_agua * 9.81  # Fuerza ejercida por la crecida

# Función para calcular el impacto de las vibraciones sísmicas
def calcular_impacto_terremoto(intensidad_terremoto, longitud_onda_terremoto):
    return intensidad_terremoto * longitud_onda_terremoto * 10000  # Fuerza sísmica generada (simplificación)

# Cálculos de los impactos
impacto_crecida = calcular_impacto_crecida(caudal_maximo, densidad_agua)
impacto_terremoto = calcular_impacto_terremoto(intensidad_terremoto, longitud_onda_terremoto)

# Evaluación de la estabilidad estructural
total_fuerza_externa = impacto_crecida + impacto_terremoto
factores_seguridad = fuerza_maxima_resistencia / total_fuerza_externa

# Mostrar los resultados
print(f"Fuerza de presión ejercida por la crecida del río: {impacto_crecida:.2f} N")
print(f"Fuerza sísmica generada por el terremoto: {impacto_terremoto:.2f} N")
print(f"Total de la fuerza externa sobre el puente: {total_fuerza_externa:.2f} N")
print(f"Factor de seguridad: {factores_seguridad:.2f} (Cuanto mayor, más estable)")

# Graficar el impacto en la estabilidad estructural
fuerzas = [impacto_crecida, impacto_terremoto]
etiquetas = ['Crecida del Río', 'Terremoto']

plt.figure(figsize=(8, 5))
plt.bar(etiquetas, fuerzas, color=['blue', 'red'])
plt.title("Impacto de las Condiciones Adversas en la Estabilidad del Puente")
plt.ylabel("Fuerza (N)")
plt.xlabel("Condiciones Adversas")
plt.grid(True)
plt.show()
```

```python
import folium
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Coordenadas del puente
latitud = 19.29137
longitud = -103.79759

# Crear un mapa interactivo
mapa = folium.Map(location=[latitud, longitud], zoom_start=15)

# Añadir un marcador para el puente
folium.Marker(
    location=[latitud, longitud],
    popup="Puente del Arroyo Seco",
    tooltip="Puente del Arroyo Seco - Villa de Álvarez",
    icon=folium.Icon(color="blue", icon="info-sign")
).add_to(mapa)

# Añadir un área de análisis (radio de 1 km)
folium.Circle(
    radius=1000,
    location=[latitud, longitud],
    popup="Área de análisis (1 km)",
    color="red",
    fill=True,
    fill_color="red",
    fill_opacity=0.3
).add_to(mapa)

# Guardar el mapa interactivo
mapa.save("rehabilitacion_puente_arroyo_seco.html")
print("Mapa interactivo generado: rehabilitacion_puente_arroyo_seco.html")

# Datos climáticos simulados: lluvias anuales (mm)
np.random.seed(42)
meses = [
    "Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
    "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"
]
lluvias = np.random.normal(loc=120, scale=30, size=12)  # Media de 120 mm
datos_lluvias = pd.DataFrame({"Mes": meses, "Lluvia (mm)": lluvias})
print("\nDatos simulados de lluvias:")
print(datos_lluvias)

# Graficar proyección de lluvias
plt.figure(figsize=(10, 6))
plt.bar(datos_lluvias["Mes"], datos_lluvias["Lluvia (mm)"], color="blue", alpha=0.7)
plt.title("Proyección de Lluvias Anuales en la Zona del Arroyo Seco")
plt.xlabel("Meses")
plt.ylabel("Lluvia (mm)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Parámetros estructurales
longitud_puente = 40  # metros
ancho_puente = 10      # metros
altura_puente = 4     # metros
volumen_puente = longitud_puente * ancho_puente * altura_puente

# Análisis de carga por crecidas
caudal_base = 50  # m³/s
incremento_caudal = np.linspace(50, 300, 6)  # Incrementos en 50 m³/s hasta 300 m³/s

# Calcular presión ejercida por el agua sobre el puente
densidad_agua = 1000  # kg/m³
gravedad = 9.81  # m/s²
altura_crecida = incremento_caudal / (longitud_puente * ancho_puente)  # Altura simulada
presion_ejercida = densidad_agua * gravedad * altura_crecida

print("\nAnálisis de presión por crecidas:")
for i, caudal in enumerate(incremento_caudal):
    print(f"Caudal: {caudal:.1f} m³/s, Altura: {altura_crecida[i]:.2f} m, Presión: {presion_ejercida[i]:.2f} Pa")

# Gráfica de presión frente a caudales
plt.figure(figsize=(10, 6))
plt.plot(incremento_caudal, presion_ejercida, marker="o", linestyle="-", color="darkorange")
plt.title("Presión Ejercida sobre el Puente vs. Caudal")
plt.xlabel("Caudal (m³/s)")
plt.ylabel("Presión (Pa)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Simulación de estado estructural (actual vs rehabilitación)
x = np.linspace(0, longitud_puente, 100)
y = np.linspace(0, ancho_puente, 100)
X, Y = np.meshgrid(x, y)
Z_actual = altura_puente - 0.05 * (X**2 + Y**2)  # Estado actual
Z_rehabilitado = altura_puente - 0.02 * (X**2 + Y**2)  # Estado tras reforzamiento

# Gráfica 3D comparativa
fig = plt.figure(figsize=(12, 8))

# Estado actual
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z_actual, cmap='inferno', edgecolor='none')
ax1.set_title("Estado Actual del Puente")
ax1.set_xlabel("Longitud (m)")
ax1.set_ylabel("Ancho (m)")
ax1.set_zlabel("Altura (m)")

# Estado rehabilitado
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, Z_rehabilitado, cmap='viridis', edgecolor='none')
ax2.set_title("Estado Rehabilitado del Puente")
ax2.set_xlabel("Longitud (m)")
ax2.set_ylabel("Ancho (m)")
ax2.set_zlabel("Altura (m)")

plt.tight_layout()
plt.show()
```
